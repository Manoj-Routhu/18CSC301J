```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/API_EN.ATM.CO2E.PC_DS2_en_excel_v2_4770549.csv")
```

```python
data.head()
```

###Data Preprocessing

```python
dataN = data.copy()
```

```python
dataN = dataN.drop("Indicator Name", axis = 'columns')
```

```python
dataN
```

```python
dataN = dataN.drop("Indicator Code", axis = "columns")
dataN = dataN.drop("1960", axis = "columns")
dataN = dataN.drop("1961", axis = "columns")
dataN = dataN.drop("1962", axis = "columns")
dataN = dataN.drop("1963", axis = "columns")
dataN = dataN.drop("1964", axis = "columns")
dataN = dataN.drop("1965", axis = "columns")
dataN = dataN.drop("1966", axis = "columns")
```

```python
dataN = dataN.drop(["1967","1968","1969","1970","1971"], axis = "columns")
```

```python
dataN = dataN.drop(["1972","1973","1974","1975","1976","1977", "1978", "1979", "1980","1981", "1982", "1983", "1984", "1985", "1986", "1987", "1988", "1989"], axis = "columns")
```

```python
dataN = dataN.drop(["2020","2021"], axis = "columns")
```

```python
dataN = dataN.drop(["Country Name","Country Code"], axis = "columns")
```

```python
dataN.info()
```

```python
dataN.dropna(axis = 0, thresh = 20, inplace = True)
```

```python
dataN.fillna(method = "backfill", inplace = True,)
```
